```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
from scipy.optimize import curve_fit
```

```python
x = np.arange(-4,4,0.1)
y = 2*np.power(x,3) + 3*np.power(x,2) -8*x + 5
```

```python
plt.plot(x,y)
```

```python
yn = y + np.random.randn(len(y))*20
plt.plot(x,y)
plt.plot(x,yn, '.')
```

```python
p = np.polyfit(x,yn,3)
```

```python
ye = np.polyval(p, x)
```

```python
plt.plot(x,y)
plt.plot(x,yn, '.')
plt.plot(x, ye)
```

```python
def f(x, a, b):
    return (a-b*x - x**3/19)*np.sin(x**2)
```

```python
a = 4
b = 3
x = np.arange(-3,3,0.1)
y = f(x,a,b)
```

```python
xn = np.arange(-3,3,0.2)
yn = f(xn,a,b) + np.random.randn((len(xn))) * 10
```

```python
plt.plot(x,y)
plt.plot(xn, yn, '.')
```

```python
p, _ = curve_fit(f, xn, yn)
p
```

```python
plt.plot(x,y)
plt.plot(xn, yn, '.')
plt.plot(x, f(x, *p))
```
